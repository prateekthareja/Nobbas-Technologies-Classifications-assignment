**For classification based on location of the property**

```python
def location(data, country, state, city='None', street='None', postal_code='none'):
    
    """This function returns the data of properties present in location specified by the user. 
    It takes the data of listings, country and state as mandatory arguments and city, street
    and postal code as optional arguments"""
    
    #Convert the data and the given inputs to lowercase strings for easy comparison
    data = data.apply(lambda x: x.astype(str).str.lower()) #1
    country = country.lower() 
    state = state.lower()
    city = city.lower()
    street = street.lower()
    postal_code = str(postal_code)
    
    #Check the given inputs and return data accordingly.
    if ((city=='none') & (street=='none') & (postal_code=='none')):
        return data[(data['country']==country) & (data['stateOrProvince']==state)]
    
    elif ((city!='none') & (street=='none') & (postal_code=='none')):
        return data[(data['country']==country) & (data['stateOrProvince']==state) & (data['city']==city)]
    
    elif ((city!='none') & (street!='none') & (postal_code=='none')):
        return data[(data['country']==country) & (data['stateOrProvince']==state) & (data['city']==city) &
                    (data['fullStreetAddress']==street)]
    
    elif ((city!='none') & (street=='none') & (postal_code!='none')):
        return data[(data['country']==country) & (data['stateOrProvince']==state) & (data['city']==city) &
                    (data['postalCode']==postal_code)]
    
    elif ((city!='none') & (street!='none') & (postal_code!='none')):
        return data[(data['country']==country) & (data['stateOrProvince']==state) & (data['city']==city) &
                    (data['fullStreetAddress']==street) & (data['postalCode']==postal_code)]
```

```python
location(data1, country='uS', state='mo', city="O'fallon", postal_code=63366, street='27 clear meadows') #Checking the result of location function.
```

**For classification based on the listing category**

```python
def listing_category(data, category):
    
    """This function returns the properties which are either rentals or are available for purchase, based on the given input.
    It takes two mandatory inputs namely: the data of listings and the category specified by the user"""
    
    #Convert the inputs into lowercase and return the instances where the listing category matches the input category.
    category = category.lower()
    return data[data['listingCategory'].apply(lambda x: x.lower())==category]
```

```python
listing_category(data1, 'rent') #Checking the result of listing_category function.
```

**For classification based on amenities**

```python
def amenities(data, bathrooms='none', bedrooms='none', parking='none'):
    
    """This function returns the data of properties, which have amenities matching those specified by the user.
    It takes the data of listings as mandatory input. Atleast one argument out of the number of bathrooms, 
    bedrooms and parking spaces is required."""
    
    #Check the given inputs and return data accordingly
    if ((bathrooms!='none') & (bedrooms!='none') & (parking!='none')):
        return data[(data['bathrooms']==bathrooms) & (data['bedrooms']==bedrooms) & (data['numParkingSpaces']==parking)]
    
    elif ((bathrooms!='none') & (bedrooms=='none') & (parking=='none')):
        return data[data['bathrooms']==bathrooms]
    
    elif ((bathrooms=='none') & (bedrooms!='none') & (parking=='none')):
        return data[data['bedrooms']==bedrooms]
    
    elif ((bathrooms=='none') & (bedrooms=='none') & (parking!='none')):
        return data[data['numParkingSpaces']==parking]
    
    elif ((bathrooms!='none') & (bedrooms!='none') & (parking=='none')):
        return data[(data['bathrooms']==bathrooms) & (data['bedrooms']==bedrooms)]
    
    elif ((bathrooms!='none') & (bedrooms=='none') & (parking!='none')):
        return data[(data['bathrooms']==bathrooms) & (data['numParkingSpaces']==parking)]
    
    elif ((bathrooms=='none') & (bedrooms!='none') & (parking!='none')):
        return data[(data['bedrooms']==bedrooms) & (data['numParkingSpaces']==parking)]
```

```python
amenities(data1, bathrooms=3, bedrooms=4, parking=2) #Checking the result of amenities function.
```
